# Histogram
This histogram shows the distribution of how much countriesâ€™ government spends on education as a % of their GDP.

### SetUp

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Pre-Processing
1. Filter Columns
2. Rename Columns

```python
# read in data
education_expend = pd.read_csv('../data/govt_education_expenditure_of_GDP.csv')

# 1. filter columns
education_expend = education_expend[['LOCATION','Time', 'Value']]

# 2. rename column
education_expend.columns = ['LOCATION','Time', 'Expenditure%']
```

### Calculations

```python
education_expend = education_expend.set_index(["LOCATION", "Time"])

# average education expenditure by country
mean_education_expend = education_expend.groupby('LOCATION').mean()
mean_education_expend.head()
```

### Plot

```python
sns.distplot(mean_education_expend['Expenditure%'].dropna().values)

# size 
plt.gcf().set_size_inches(11.7, 8.27)

# title
plt.title(f'Goverentment Expenditure Around the World',fontsize=20)

#axis
plt.xlabel('Education Expenditure as a % of GDP', fontsize=18)
plt.ylabel('Frequency Density', fontsize=18)
plt.tick_params(labelsize=14)

# plt.show()
plt.savefig('Histogram')
```
